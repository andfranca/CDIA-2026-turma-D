#### Enviar Formulário


```python
# Importar bibliotecas
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.keys import Keys
```

```python
# Endereço da página
url = 'https://forms.gle/matymZzz5ipXKhfh8'
```

```python
# Executar o webdriver
service = Service(ChromeDriverManager().install())
driver = webdriver.Chrome(service=service)
```

```python
import time

# acessar a página
driver.get(url)

# Encontrar campos
nome = driver.find_element(By.CLASS_NAME,'whsOnd.zHQkBf')
texto = driver.find_element(By.CLASS_NAME,'KHxj8b.tL9Q4c')
selecao = driver.find_element(By.ID,'i12') #escolher entre i15, i18 ou i21
opcoe = driver.find_element(By.ID,'i26') #Escolher entre i26, i29 ou i32

# Preencher com as informações desejadas
nome.clear()
nome.send_keys('João')

selecao.click()
opcoe.click()

texto.clear()
texto.send_keys('Isso é um texto longo')

time.sleep(1)

#Clicar em enviar
botao_enviar = driver.find_element(By.CLASS_NAME,'NPEfkd.RveJvd.snByac')
botao_enviar.click()

time.sleep(2)

driver.get(url)

```

#### Preenchimento automatizado

```python
import time
import pandas as pd

from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

df = pd.read_csv("dados.csv")

for _, linha in df.iterrows():

    driver.get(url)

    espera = WebDriverWait(driver, 10)

    # Campos do formulário
    nome = espera.until(
        EC.element_to_be_clickable(
            (By.CSS_SELECTOR, ".whsOnd.zHQkBf")
        )
    )

    texto = driver.find_element(
        By.CSS_SELECTOR, ".KHxj8b.tL9Q4c"
    )

    # IDs disponíveis para cada alternativa
    selecoes = {
        1: "i15",
        2: "i18",
        3: "i21"
    }

    opcoes = {
        1: "i26",
        2: "i29",
        3: "i32"
    }

    # Preencher o formulário
    nome.send_keys(str(linha["nome"]))

    driver.find_element(
        By.ID,
        selecoes[int(linha["selecao"])]
    ).click()

    driver.find_element(
        By.ID,
        opcoes[int(linha["opcao"])]
    ).click()

    texto.send_keys(str(linha["texto"]))

    # Enviar
    botao_enviar = driver.find_element(
        By.CSS_SELECTOR, ".NPEfkd.RveJvd.snByac"
    )

    botao_enviar.click()

    time.sleep(2)
```
