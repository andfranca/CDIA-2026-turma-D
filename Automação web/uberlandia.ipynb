```python
import pandas as pd

# Importar bibliotecas
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.keys import Keys
```

```python
url = 'https://www.uberlandia.mg.gov.br/prefeitura/orgaos-municipais/procuradoria-geral-do-municipio/diario-oficial-uberlandia/'
```

```python
# Executar o webdriver
service = Service(ChromeDriverManager().install())
driver = webdriver.Chrome(service=service)

driver.get(url)
```

```python
# visualizar nome de últimas publicações
ultimas_publicacoes = driver.find_elements(By.CLASS_NAME,'ee-post__title__heading')
data = driver.find_elements(By.CLASS_NAME,'ee-post__metas__list.display--')

driver.quit()
```

```python
#Dicionário de meses
meses = {
    "janeiro": "01",
    "fevereiro": "02",
    "março": "03",
    "abril": "04",
    "maio": "05",
    "junho": "06",
    "julho": "07",
    "agosto": "08",
    "setembro": "09",
    "outubro": "10",
    "novembro": "11",
    "dezembro": "12"
}
```

```python
#Pegar as últimas publicações
publicacoes = {}

for i in range(len(ultimas_publicacoes)):
    publicacoes[ultimas_publicacoes[i].text] = data[i].text

  
```

```python
#montar os links com as publicações
links = []


for edicao, data in publicacoes.items():

    numero = edicao.split()[-1]

    partes_data = data.split()
    mes = meses[partes_data[2]]
    ano = partes_data[4]

    link = (
        f"https://docs.uberlandia.mg.gov.br/"
        f"wp-content/uploads/{ano}/{mes}/{numero}.pdf"
    )

    links.append(link)
    
```

```python
#download do arquivo pdf 
import os
import requests

os.makedirs("pdf_uberlandia", exist_ok=True)

for link in links:
    nome_arquivo = link.split("/")[-1]
    caminho = os.path.join("arquivos", nome_arquivo)

    response = requests.get(link)
    response.raise_for_status()

    with open(caminho, "wb") as f:
        f.write(response.content)
```
