```python
import pandas as pd
import os
import time
# Importar bibliotecas
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.keys import Keys
```

```python
pasta_downloads = os.path.abspath("dados_vacina")

os.makedirs(pasta_downloads, exist_ok=True)
```

```python
opcoes = webdriver.ChromeOptions()

opcoes.add_experimental_option(
    "prefs",
    {
        "download.default_directory": pasta_downloads,
        "download.prompt_for_download": False,
        "download.directory_upgrade": True,
        "plugins.always_open_pdf_externally": True
    }
)
```

```python
url = 'https://infoms.saude.gov.br/extensions/SEIDIGI_DEMAS_ESTRATEGIA_INFLUENZA_RESIDENCIA/index.html?regiao=nacional'
```

```python
# Executar o webdriver
service = Service(ChromeDriverManager().install())
driver = webdriver.Chrome(service=service,options=opcoes)

driver.get(url)

time.sleep(15)
```

```python
driver.find_element(By.ID, 'home-tab').click()
```

```python
driver.find_element(By.CSS_SELECTOR,'[aria-label="Cód Município Residência"]').click()
time.sleep(1)

driver.find_element(By.CSS_SELECTOR,'[data-testid="search-input-field"]').send_keys('41370 - Londrina')
time.sleep(2)

driver.find_element(By.CLASS_NAME,'RowColumn-barContainer.css-1odcosj').click()
time.sleep(1)
```

```python
driver.find_element(By.CSS_SELECTOR,'[title="Confirmar seleção"]').click()
time.sleep(3)
```

```python
botao_download = WebDriverWait(driver, 10).until(
    EC.element_to_be_clickable((By.ID, "exportar_dados_grafico_03"))
)

driver.execute_script("arguments[0].click();", botao_download)
```
