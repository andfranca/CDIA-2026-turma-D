```python
import pymupdf4llm
```

```python
caminho = 'nota.pdf'

md = pymupdf4llm.to_markdown(caminho)
```

```python
print(md)
```

```python
with open('documento_markdown.md', 'w', encoding='utf-8') as f:
    f.write(md)
```

```python
import ollama

resposta = ollama.chat(
    model="gemma4:e2b",
    messages=[
        {
        "role": "user", 
        "content": f""" 
        
        Você é um assistente de IA especializado em análise de alterações contratuais
        Seja direto e objetivo em suas respostas        
        Analise a nota fiscal extraída de um PDF e extraia as seguintes informações:
        
        Identificação do prestador
        Razão social do prestador, 
        CNPJ do prestador, 
        numero do contrato,
        Situação do prestador
        
        
        Informações sobre as alíquotas
        
       
        Identificação da empresa
        coloque as iformações em formato de tabela, com as seguintes colunas:
        Razão social do prestador, CNPJ do prestador, numero do contrato, Situação do prestador, Alíquotas
        
         "{md}"
         """        
         
         },
    ],
)

print(resposta["message"]["content"])
```
