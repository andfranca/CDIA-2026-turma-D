```python
import requests
from bs4 import BeautifulSoup
```

```python
url = "https://www.gov.br/anvisa/pt-br/assuntos/noticias-anvisa/2026/veja-os-destaques-da-10a-reuniao-ordinaria-publica-da-diretoria-colegiada"
```

```python
resposta = requests.get(url)
soup = BeautifulSoup(resposta.text, 'html.parser')
```

```python
titulo = soup.find('h1')
descricao = soup.find('div', class_='documentDescription')

print(titulo.text.strip())
print(descricao.text.strip())
```

```python
paragrafos = soup.find_all('p')

for p in paragrafos:
    print(p.text.strip())
```

```python
import tiktoken
from bs4 import BeautifulSoup

caminho = "noticia.html"

# Lê o conteúdo do arquivo HTML
with open(caminho, "r", encoding="utf-8") as f:
    html = f.read()

# Parser com BeautifulSoup
soup = BeautifulSoup(html, "html.parser")

# Extrai todo o texto do body
body = soup.find("body")

if body:
    texto_limpo = body.get_text(separator=" ", strip=True)
else:
    texto_limpo = soup.get_text(separator=" ", strip=True)

print(texto_limpo)

# Tokenizador
encoding = tiktoken.encoding_for_model("gpt-4")

tokens_html = encoding.encode(html)
print(f"Tokens HTML: {len(tokens_html)}")

tokens_texto_limpo = encoding.encode(texto_limpo)
print(f"Tokens Texto Limpo: {len(tokens_texto_limpo)}")

# Preços por token
preco_input = 2 / 1000000
preco_output = 8 / 1000000

# Exemplo de cálculo
custo_estimado = (
    preco_input * len(tokens_texto_limpo)
    + preco_output * len(tokens_texto_limpo)
)

# Exemplo de cálculo
custo_estimado2 = (
    preco_input * len(tokens_html)
    + preco_output * len(tokens_html)
)


print(f"Custo estimado: US$ {custo_estimado:.6f}")
print(f"Custo estimado (HTML): US$ {custo_estimado2:.6f}")
```
